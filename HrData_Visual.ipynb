```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\D_Drive\Reshma\KP_Study\hr_data.xlsx"
df = pd.read_excel(file_path)

# Convert 'Date' column to datetime format (if not already)
df["Date of Joining"] = pd.to_datetime(df["Date of Joining"])

# Display first few rows
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Sales Trend Over Time (Line Chart) 
df["Date of Joining"] = pd.to_datetime(df["Date of Joining"])  # Ensure Date column is in datetime format
df_resampled = df.resample("M", on="Date of Joining").sum()  # Resample by month

plt.figure(figsize=(8, 5))
plt.plot(df_resampled.index, df_resampled["Salary ($)"], marker="o", linestyle="-", color="b")
plt.xlabel("Date of Joining")
plt.ylabel("Salary ($)")
plt.title("Salary Trend Over Time (Monthly)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Salary by Category (Bar Chart) 
# Use seaborn and matplotlib in combination 
plt.figure(figsize=(6, 4))
sns.barplot(x="Department", y="Salary ($)", data=df, palette=["red", "blue", "green"])
plt.title("Salary by Department")
plt.xticks(rotation=45)
plt.show()
```

```python
# Revenue Distribution by Region (Pie Chart) 
df.info()
```

```python
df_grouped = df.groupby("Region")["Salary ($)"].sum()
plt.pie(df_grouped, labels=df_grouped.index, autopct="%1.1f%%", colors=plt.cm.Paired.colors)
plt.show()
```

```python
# Salary vs. Department (Scatter Plot)
plt.figure(figsize=(6, 4))
sns.scatterplot(x="Experience (Years)", y="Salary ($)", data=df)
plt.title("Salary vs. Department")
plt.show()
```

```python
# Product-Wise Revenue (Horizontal Bar Chart)
plt.figure(figsize=(6, 4))
sns.barplot(x="Salary ($)", y="Department", data=df, ci=None, palette=["red", "blue", "green", "purple", "orange"])
plt.title("Department-Wise Salary")
plt.show()
```
