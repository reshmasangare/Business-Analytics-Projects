```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = pd.read_excel(r'C:\D_Drive\Reshma\EmpData.xlsx')
df.head()
```

```python
# Defining independent and dependent variables

x = df.drop(columns=["Attrition"]) # independent variable
y = df["Attrition"] # dependent variable

#x = df.drop(columns=["CTCoffered"]) #independent variable
#y = df["CTCoffered"] #dependent variable
```

```python
x.head()
```

```python
y.head()
```

```python
#Splitting data into test and train datasets [20-80]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 4)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
#building model

model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# Predicting based on Test Data
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# Performance Metrics
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
model.coef_
```

```python
model.intercept_
```
